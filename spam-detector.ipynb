```python
# Step 1: Libraries import karo
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# Step 2: Data load karo - ye dataset free milta hai
url = 'https://raw.githubusercontent.com/justmarkham/pycon-2016-tutorial/master/data/sms.tsv'
df = pd.read_csv(url, sep='\t', header=None, names=['label', 'message'])

# Step 3: Data ko numbers me badlo taake model samjhe
df['label'] = df.label.map({'ham': 0, 'spam': 1})
X_train, X_test, y_train, y_test = train_test_split(df['message'], df['label'], test_size=0.2)

vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train)
X_test_count = vectorizer.transform(X_test)

# Step 4: Model train karo - ye AI ka magic hai
model = MultinomialNB()
model.fit(X_train_count, y_train)

# Step 5: Test karo kitna accurate hai
predictions = model.predict(X_test_count)
print("Accuracy:", accuracy_score(y_test, predictions))

# Step 6: Apna message check karo
my_message = ["Win a FREE iPhone now!!! Click here"]
my_message_count = vectorizer.transform(my_message)
result = model.predict(my_message_count)
print("Tumhara message:", "Spam" if result[0]==1 else "Not Spam")
```
